## The Python web scraping notebook 3: Player ODI Rankings

### https://www.icc-cricket.com/
* Mens cricket player ODI Rankings for Batsmen, Bowlers and All-Rounders

```python
# Importing the libraries
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

### Getting the Latest Team rankings

```python
# Test links for webscraping
odibat = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
odibowl = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"
odiallrounder = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/all-rounder"
```

### Scraping the ODI batting data

```python
# The odi batting data
response = requests.get(odibat)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")
```

```python
# For ODI Player Batting Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


odi_bat_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })
odi_bat_rank['Type'] = 'Batting'
odi_bat_rank['Format'] = 'ODI'
odi_bat_rank
```

### Scraping the ODI bowling data

```python
# The odi bowling data
response = requests.get(odibowl)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")

# For ODI Player Bowling Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


odi_bowl_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })

odi_bowl_rank['Type'] = 'Bowling'
odi_bowl_rank['Format'] = 'odi'

odi_bowl_rank
```

### Scraping the odi all-rounder data

```python
# The odi bowling data
response = requests.get(odiallrounder)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")

# For ODI Player All rounder Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


odi_allrounder_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })

odi_allrounder_rank['Type'] = 'All-rounder'
odi_allrounder_rank['Format'] = 'odi'

odi_allrounder_rank
```

```python
#### Overall ODI Ranking Data

Ranking_data = pd.concat([odi_bat_rank, odi_bowl_rank, odi_allrounder_rank], axis=0, ignore_index = True)
Ranking_data
```

```python
# Writing out

Ranking_data.to_csv("Player_ODI_Rankings.csv",index=False)
```
