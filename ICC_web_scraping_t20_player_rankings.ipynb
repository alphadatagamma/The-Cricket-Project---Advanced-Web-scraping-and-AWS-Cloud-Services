## The Python web scraping notebook 4: Player T20 Rankings

### https://www.icc-cricket.com/
* Mens cricket player T20 Rankings for Batsmen, Bowlers and All-Rounders

```python
# Importing the libraries
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

### Getting the Latest Team rankings

```python
# T20i links for webscraping
t20bat = "https://www.icc-cricket.com/rankings/mens/player-rankings/t20i/batting"
t20bowl = "https://www.icc-cricket.com/rankings/mens/player-rankings/t20i/bowling"
t20allrounder = "https://www.icc-cricket.com/rankings/mens/player-rankings/t20i/all-rounder"
```

### Scraping the test batting data

```python
# The t20i batting data
response = requests.get(t20bat)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")
```

```python
# For T20 Player Batting Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


t20_bat_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })
#test_rank
```

```python
t20_bat_rank['Type'] = 'Batting'
t20_bat_rank['Format'] = 'T20'

t20_bat_rank
```

### Scraping the test bowling data

```python
# The t20 bowling data
response = requests.get(t20bowl)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")

# For T20 Player Bowling Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


t20_bowl_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })

t20_bowl_rank['Type'] = 'Bowling'
t20_bowl_rank['Format'] = 'T20'

t20_bowl_rank
```

### Scraping the test all-rounder data

```python
# The t20 all rounder data
response = requests.get(t20allrounder)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")

# For T20 Player All rounder Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


t20_allrounder_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })

t20_allrounder_rank['Type'] = 'All-rounder'
t20_allrounder_rank['Format'] = 'T20'

t20_allrounder_rank
```

```python
#### Overall T20_Ranking Data

Ranking_data = pd.concat([t20_bat_rank, t20_bowl_rank, t20_allrounder_rank], axis=0, ignore_index = True)
Ranking_data
```

```python
# Writing out

Ranking_data.to_csv("Player_T20_Rankings.csv",index=False)
```
