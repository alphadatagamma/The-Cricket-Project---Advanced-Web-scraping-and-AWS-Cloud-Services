## The Python web scraping notebook 2: Player Test Rankings

### https://www.icc-cricket.com/
* Mens cricket player Test Rankings for Batsmen, Bowlers and All-Rounders

```python
# Importing the libraries
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

### Getting the Latest Team rankings

```python
# Test links for webscraping
testbat = "https://www.icc-cricket.com/rankings/mens/player-rankings/test/batting"
testbowl = "https://www.icc-cricket.com/rankings/mens/player-rankings/test/bowling"
testallrounder = "https://www.icc-cricket.com/rankings/mens/player-rankings/test/all-rounder"
```

### Scraping the test batting data

```python
# The test batting data
response = requests.get(testbat)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")
```

```python
# For Test Player Batting Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


test_bat_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })
#test_rank
```

```python
test_bat_rank['Type'] = 'Batting'
test_bat_rank['Format'] = 'Test'

test_bat_rank
```

### Scraping the test bowling data

```python
# The test bowling data
response = requests.get(testbowl)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")

# For Test Player Bowling Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

#print(len(player_name))
#print(len(player_country))
#print(len(player_rating))


test_bowl_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })

test_bowl_rank['Type'] = 'Bowling'
test_bowl_rank['Format'] = 'Test'

test_bowl_rank
```

### Scraping the test all-rounder data

```python
# The test bowling data
response = requests.get(testallrounder)
html_soup = BeautifulSoup(response.text, 'html.parser')
text = str(html_soup.find_all("td"))
#html_soup.find_all("td")

# For Test Player All rounder Ranks

player_name = re.findall(r'(?<="\>)[A-Za-z -.]+(?=<\/a)', text)
player_name = re.findall(r'(?<=large"\>)[A-Za-z ]+', text) + player_name
player_country = re.findall(r'(?<=text">)[A-Za-z ]+(?=<)',text)
player_country = re.findall(r'(?<=span>)[A-Za-z ]+', text) + player_country
player_rating = re.findall(r'(?<=rating">)[0-9]+(?=<)',text)

print(len(player_name))
print(len(player_country))
print(len(player_rating))


test_allrounder_rank = pd.DataFrame({'Rank' : range(len(player_name)),
                            'Player Name' : player_name,
                              'Player Country' : player_country,
                                      'Ratings': player_rating
                                })

test_allrounder_rank['Type'] = 'All-rounder'
test_allrounder_rank['Format'] = 'Test'

test_allrounder_rank
```

```python
#### Overall Test_Ranking Data

Ranking_data = pd.concat([test_bat_rank, test_bowl_rank, test_allrounder_rank], axis=0, ignore_index = True)
Ranking_data
```

```python
# Writing out

Ranking_data.to_csv("Player_Test_Rankings.csv",index=False)
```
