## The Python web scraping notebook 1: Team rankings

### https://www.icc-cricket.com/
* Mens cricket Test, ODI and T-20 Team rankings 

```python
# Importing the libraries
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

### Getting the Latest Team rankings

```python
# The links for webscraping
testlink = "https://www.icc-cricket.com/rankings/mens/team-rankings/test"
odilink = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
t20link = "https://www.icc-cricket.com/rankings/mens/team-rankings/t20i"
```

### Scraping the test data

```python
# The test data
response = get(testlink)
html_soup = BeautifulSoup(response.text, 'html.parser')
country_text = str(html_soup.find_all('span', class_ = "u-hide-phablet"))
ctry_text = str(html_soup.find_all('span', class_ = "u-show-phablet"))
country_list = re.findall(r'(?<=[\>])[A-Za-z ]+(?=<)',country_text)
ctry_list = re.findall(r'(?<=[\>])[A-Za-z ]+(?=<)',ctry_text)
#ctry_list
#country_list
detail_text = str(html_soup.find_all('table'))
detail_text2 = re.sub(r'\s+',"",detail_text)
#detail_text2 
num_lst= re.findall(r'(?<="[\>])[0-9,]+',detail_text2)
#num_lst
rank = num_lst[0::4]
matches = num_lst[1::4]
points = num_lst[2::4]
rating = num_lst[3::4]

test_rank = pd.DataFrame({'Rank' : rank,
                            'Country' : ctry_list,
                                'Country Name' : country_list,
                                  'Matches':matches,
                                      'Points': points,
                                      'Ratings': rating
                                })
```

```python
test_rank['Format'] = 'Test'
test_rank
```

```python
# The odi data

response = get(odilink)
html_soup = BeautifulSoup(response.text, 'html.parser')
country_text = str(html_soup.find_all('span', class_ = "u-hide-phablet"))
ctry_text = str(html_soup.find_all('span', class_ = "u-show-phablet"))
country_list = re.findall(r'(?<=[\>])[A-Za-z ]+(?=<)',country_text)
ctry_list = re.findall(r'(?<=[\>])[A-Za-z ]+(?=<)',ctry_text)
#ctry_list
#country_list
detail_text = str(html_soup.find_all('table'))
detail_text2 = re.sub(r'\s+',"",detail_text)
#detail_text2 
num_lst= re.findall(r'(?<="[\>])[0-9,]+',detail_text2)
#num_lst
rank = num_lst[0::4]
matches = num_lst[1::4]
points = num_lst[2::4]
rating = num_lst[3::4]

odi_rank = pd.DataFrame({'Rank' : rank,
                            'Country' : ctry_list,
                                'Country Name' : country_list,
                                  'Matches':matches,
                                      'Points': points,
                                      'Ratings': rating
                                })
odi_rank['Format'] = 'ODI'
odi_rank
```

```python
# The t20 data
response = get(t20link)
html_soup = BeautifulSoup(response.text, 'html.parser')
country_text = str(html_soup.find_all('span', class_ = "u-hide-phablet"))
ctry_text = str(html_soup.find_all('span', class_ = "u-show-phablet"))
country_list = re.findall(r'(?<=[\>])[A-Za-z ]+(?=<)',country_text)
ctry_list = re.findall(r'(?<=[\>])[A-Za-z ]+(?=<)',ctry_text)
#ctry_list
#country_list
detail_text = str(html_soup.find_all('table'))
detail_text2 = re.sub(r'\s+',"",detail_text)
#detail_text2 
num_lst= re.findall(r'(?<="[\>])[0-9,]+',detail_text2)
#num_lst
rank = num_lst[0::4]
matches = num_lst[1::4]
points = num_lst[2::4]
rating = num_lst[3::4]

t20_rank = pd.DataFrame({'Rank' : rank,
                            'Country' : ctry_list,
                                'Country Name' : country_list,
                                  'Matches':matches,
                                      'Points': points,
                                      'Ratings': rating
                                })
t20_rank['Format'] = 'T20'
t20_rank
```

```python
#### Overall Ranking Data

Ranking_data = pd.concat([test_rank, odi_rank, t20_rank], axis=0, ignore_index = True)
Ranking_data
```

```python
# Writing out

Ranking_data.to_csv("Team_Rankings.csv",index=False)
```
